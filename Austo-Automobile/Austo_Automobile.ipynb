The Gender starts with F is considered as Female

```python
# Function to standardize gender based on the first letter
def standardize_gender(gender):
    if isinstance(gender, str):  # Check if the value is a string
        gender_lower = gender.lower()
        if gender_lower.startswith('f'):
            return 'Female'
        # You can add more conditions here if needed,
        # for example, to standardize 'M' to 'Male'
        # elif gender_lower.startswith('m'):
        #     return 'Male'
        else:
            return gender  # Keep the original value if it doesn't start with 'F'
    else:
        return gender  # Keep non-string values as they are (e.g., NaN)

# Apply the function to the 'Gender' column
df['Gender'] = df['Gender'].apply(standardize_gender)

# Check the unique values after standardization to see the effect
print(df['Gender'].unique())
```

Since Total salary = customer salary + Partner's salary, will impute the mi9ssing valus with this formula

```python
# Identify rows where Partner_salary is missing
missing_partner_indices = df[df['Partner_salary'].isna()].index
```

```python
# Impute Partner_salary based on the Total_salary - Salary for those missing entries
for index in missing_partner_indices:
    total_salary = df.at[index, 'Total_salary']
    salary = df.at[index, 'Salary']
    # Calculate Partner_salary
    partner_salary = total_salary - salary

    # Ensure the calculated partner_salary is not negative before imputing
    if partner_salary >= 0:
        df.at[index, 'Partner_salary'] = partner_salary
```

Outliers Treat for Total salary using IQR and clipping

```python
# Outlier treatment example using IQR for Total Salary
Q1 = df['Total_salary'].quantile(0.25)
Q3 = df['Total_salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
# Fill or cap the outliers
df['Total_salary'] = np.clip(df['Total_salary'], lower_bound, upper_bound)
```

```python
# Boxplots for after treating Total Salary outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=df['Total_salary'])
plt.title('Boxplots of Total Salary after treating Outliers')
plt.show()
```

```python
suv_prices = df[df['Make'] == 'SUV']['Price']
```

6. How does having a working partner influence the purchase of higher-priced cars?

```python
with_partner = df[df['Partner_working'] == 'Yes']
without_partner = df[df['Partner_working'] == 'No']
```

```python
# Higher-priced purchases for individuals with a working partner
higher_priced_with_partner = with_partner[with_partner['Price'] > price_threshold]
higher_priced_count_with_partner = higher_priced_with_partner.shape[0]
average_spent_with_partner = higher_priced_with_partner['Price'].mean()

# Higher-priced purchases for individuals without a working partner
higher_priced_without_partner = without_partner[without_partner['Price'] > price_threshold]
higher_priced_count_without_partner = higher_priced_without_partner.shape[0]
average_spent_without_partner = higher_priced_without_partner['Price'].mean()

# Summary of the findings
print(f"Number of Higher-Priced Cars Purchased with Partner: {higher_priced_count_with_partner}")
print(f"Average Spending with Partner: ${average_spent_with_partner:.2f}")

print(f"Number of Higher-Priced Cars Purchased without Partner: {higher_priced_count_without_partner}")
print(f"Average Spending without Partner: ${average_spent_without_partner:.2f}")
```

```python
# Comparative Analysis
if higher_priced_count_with_partner > higher_priced_count_without_partner:
    print("Individuals with working partners tend to purchase more higher-priced cars.")
else:
    print("Individuals without working partners tend to purchase more higher-priced cars.")

if average_spent_with_partner > average_spent_without_partner:
    print("Overall spending on higher-priced cars is greater among individuals with working partners.")
else:
    print("Overall spending on higher-priced cars is greater among individuals without working partners.")
```

From the above observation, we can say that the customers with working partner purchase more higher priced cars.
